```python
import requests

URL = "http://vlm:7860/predict"



params = {
    "images": "1.png, 10.png",
    "prompt": "describe",
    "bucket": "color-mmls",
    "folder": "munsell_colors"
}

response = requests.get(URL, params=params)

if response.status_code == 200:
    print("Response:", response.json())
else:
    print("Error:", response.status_code, response.text)
```

```python
def ask_vlm(images: str, prompt: str):
    params = {
        "images": images,
        "prompt": prompt,
        "bucket": "color-mmls",
        "folder": "munsell_colors"
    }
    
    response = requests.get(URL, params=params)
    if response.status_code == 200:
        return response.json()
    else:
        return response.status_code, response.text
```

```python
ask_vlm("1.png, 15.png, 10.png", "Compare image 1 and image 15 with a image 10 in terms of color similarity or differences in RGB vector values")
```

```python
ask_vlm("1.png", "What color is it?")
```

```python
ask_vlm("15.png", "What color is it?")
```

```python
ask_vlm("10.png", "What color is it?")
```

```python
ask_vlm("15.png", "What color is it? Get RGB vector value")
```
